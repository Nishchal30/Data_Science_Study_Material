```python
df1 = pd.DataFrame(np.random.randint(1,100, size=(100, 2)), columns=["A", "B"])
df1
```

```python
df1.iplot()
```

iplot() basically plots the row number wise its value on the plot. 
\
exa:- row 0 has 92 value in column 1 & 34 value in column 2.

x- axis will represent the row indexes and y-axis will represent the row values.

```python
df1.iplot(x= 'A', y = 'B')
```

This graph will gives us the relationship or the trend between A & B column of df1.

```python
df1.iplot(x= 'A', y = 'B', mode="markers")
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
print(x)
print(y)
```

### **Scatter plot**

```python
plt.scatter(x, y, c='violet')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot")
plt.savefig("test.jpg")
```

### **Line plot**

```python
y = x*x
plt.plot(x, y, 'ro', linestyle= "dashed", linewidth = 3, markersize = 10) # third parameter is by default color 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line plot")
```

### **Subplots**

plt.subplot(2,2,1) means total there are 2 rows and 2 columns and 1 means at 1st position.

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'ro')
plt.subplot(2,2,2)
plt.plot(x, y, 'y--')
plt.subplot(2,2,3)
plt.plot(x, y, 'b*')
plt.subplot(2,2,4)
plt.plot(x, y, 'g^')
```
